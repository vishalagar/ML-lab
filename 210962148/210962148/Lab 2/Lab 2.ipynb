```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import pandas as pd
```

```python
x = np.arange(0,100)
y = np.arange(0,100) *2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")
plt.locator_params(axis = 'y', nbins = 5)
plt.xlim(0,max(x))
plt.ylim(0,max(y))
```

```python
x = np.arange(120)
y = np.arange(120)**2

fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
ax2 = fig1.add_axes([.2, .5, .2, .2])
ax1.plot(x,y)
ax2.plot(y,x)
```
